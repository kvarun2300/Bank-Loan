###Total Loan Applications


```python
total_loan_applications= df['id'].count()
print("Total Loan Applications:", total_loan_applications)
```

###Month To Date TLA

```python
latest_issue_date = df['issue_date'].max()
latest_year = latest_issue_date.year
latest_month = latest_issue_date.month

mtd_data = df[(df['issue_date'].dt.year == latest_year) & (df['issue_date'].dt.month == latest_month) ]

mtd_loan_applications = mtd_data['id'].count()

print(f"MTD Loan Applications(for {latest_issue_date.strftime('%B %Y')}): {mtd_loan_applications} ")
```

###Total Funded Amount

```python
total_funded_amount = df['loan_amount'].sum()
total_funded_amount_millions = total_funded_amount/1000000
print("Total Funded Amount: ${:.2f}M".format(total_funded_amount_millions))
```

###MTD Total Funded Amount

```python
latest_issue_date = df['issue_date'].max()
latest_year = latest_issue_date.year
latest_month = latest_issue_date.month

mtd_data= df[(df['issue_date'].dt.year == latest_year) & (df['issue_date'].dt.month==latest_month)]

mtd_total_funded_amount = mtd_data['loan_amount'].sum()
mtd_total_funded_amount_millions = mtd_total_funded_amount/1000000

print("MTD Total Funded A mount: ${:.2f}M".format(mtd_total_funded_amount_millions))
```

###Total Amount Recieved

```python
total_amount_recieved = df['total_payment'].sum()
total_amount_recieved_millions = total_amount_recieved/1000000
print("Total Amount Recieved: ${:.2f}M". format(total_amount_recieved_millions))
```

###MTD Total Amount Recieved

```python
latest_issue_date = df['issue_date'].max()
latest_year = latest_issue_date.year
latest_month = latest_issue_date.month

mtd_data = df[(df['issue_date'].dt.year == latest_year) & (df['issue_date'].dt.month == latest_month)]

mtd_total_amount_recieved = mtd_data['total_payment'].sum()
mtd_total_amount_recieved_millions = mtd_total_amount_recieved/1000000

print("MTD Total Amount Recieved: ${:.2f}M".format(mtd_total_amount_recieved_millions))
```

###Average Intrest Rate

```python
average_intrest_rate = df['int_rate'].mean()*100
print("Averege Intrest Rate: {:.2f}%".format(average_intrest_rate))
```

###Average Debt to Income Ratio (DTI)

```python
average_dti = df['dti'].mean()*100
print("Averege DTI: {:.2f}%".format(average_dti))
```

###Good Loan Metrics

```python
good_loan =df[df['loan_status'].isin(["Fully Paid", "Current"])]

good_loan_applications = good_loan['id'].count()
good_loan_funded = good_loan['loan_amount'].sum()
good_loan_recieved = good_loan['total_payment'].sum()

good_loan_funded_millions = good_loan_funded / 1000000
good_loan_recieved_millions = good_loan_recieved / 1000000

good_loan_percentage = (good_loan_applications/ total_loan_applications) * 100

print("Good Loan Applications:",good_loan_applications)
print("Good Loan Funded Amount: ${:.2f}M".format(good_loan_funded_millions))
print("Good Loan Recieved Amount: ${:.2f}M".format(good_loan_recieved_millions))
print("Percentage of Goad Loan Applications: {:.2f}%".format(good_loan_percentage))
```

###Bad Loan Metrics

```python
bad_loan =df[df['loan_status'].isin(["Charged Off"])]

bad_loan_applications = bad_loan['id'].count()
bad_loan_funded = bad_loan['loan_amount'].sum()
bad_loan_recieved = bad_loan['total_payment'].sum()

bad_loan_funded_millions = bad_loan_funded / 1000000
bad_loan_recieved_millions = bad_loan_recieved / 1000000

bad_loan_percentage = (bad_loan_applications/ total_loan_applications) * 100

print("Bad Loan Applications:",bad_loan_applications)
print("Bad Loan Funded Amount: ${:.2f}M".format(bad_loan_funded_millions))
print("Bad Loan Recieved Amount: ${:.2f}M".format(bad_loan_recieved_millions))
print("Percentage of Bad Loan Applications: {:.2f}%".format(bad_loan_percentage))
```
